## 1- Write mypmwrite and mypgmread functions using Image_analyzer()

```python
import numpy as np
from image_analyzer import Image_analyzer
import os

def mypgmwrite(row, column, max_pixel_value, test_image_array, dir):
    analyzer = Image_analyzer()
    analyzer.image_write(row, column, max_pixel_value, test_image_array, dir)
    print('Image written to test.pgm')
    return
def mypgmread(dir):
    analyzer = Image_analyzer()
    image_array, width, height, max_pixel_value = analyzer.image_read(dir)
    return image_array, width, height, max_pixel_value                           
```

## 2- Create a two dimentional array to write it to test.pgm file

```python
#get file directory 
file_dir = os.getcwd()
#create tow dimentional image array
test_image_array = np.array([[101,102,103],[104,105,106],[107,108,109]], dtype=np.uint8) # datatype nedded to set to 8bit
row,column = 3,3
max_pixel_value = 255
test_dir = file_dir+'/'+'test.pgm'
mypgmwrite(row, column, max_pixel_value, test_image_array, test_dir)
```

## 3- Read test.pgm file

```python
image_array, width, height, max_pixel_value = mypgmread(test_dir)
print('Image array:\n', image_array)
```
