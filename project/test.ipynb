```python
import cv2
import matplotlib.pyplot as plt

# Load the face and eye Haarcascade files
face_cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
eye_cascade_path = cv2.data.haarcascades + 'haarcascade_eye.xml'

face_cascade = cv2.CascadeClassifier(face_cascade_path)
eye_cascade = cv2.CascadeClassifier(eye_cascade_path)

# Load the image as grayscale
img_path = 'Test_images/man.jpg'  # Replace with the correct image path
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print(f"Error: Image not found at {img_path}")
else:
    # Detect faces
    faces = face_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=4)

    # For each detected face, detect eyes within the face region
    for (x, y, w, h) in faces:
        # Draw rectangle around the face
        cv2.rectangle(img, (x, y), (x+w, y+h), 255, 2)

        # Focus on the face region for eye detection
        face_region = img[y:y+h, x:x+w]

        # Detect eyes in the face region
        eyes = eye_cascade.detectMultiScale(face_region, scaleFactor=1.1, minNeighbors=10)

        for (ex, ey, ew, eh) in eyes:
            # Draw rectangle around the eyes
            cv2.rectangle(face_region, (ex, ey), (ex+ew, ey+eh), 255, 2)

    # Show the result
    plt.imshow(img, cmap='gray')
    plt.axis('off')  # Turn off axis for better display
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image as grayscale
img_path = 'Test_images/man.jpg'  # Replace with the correct image path
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print(f"Error: Image not found at {img_path}")
else:
    # Apply Gaussian blur to reduce noise
    blurred_img = cv2.GaussianBlur(img, (15, 15), 0)
    
    #edge detection using canny
    edges = cv2.Canny(blurred_img, 100, 200)
    
    #Thresholding the detected edges
    tresholded_binary = cv2.threshold(edges, 100, 255, cv2.THRESH_BINARY)[1]
    

    # Detect circles using HoughCircles
    circles = cv2.HoughCircles(tresholded_binary, 
                               cv2.HOUGH_GRADIENT, dp=1.2, minDist=30,
                               param1=50, param2=30, minRadius=10, maxRadius=100)

    # If circles are detected, draw them on the original image
    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")  # Round coordinates and convert to integers

        # Draw each circle and its center
        for (x, y, r) in circles:
            # Draw the outer circle (green)
            cv2.circle(img, (x, y), r, (0, 255, 0), 4)
            # Draw the center of the circle (red)
            cv2.circle(img, (x, y), 2, (0, 0, 255), 3)

        # Show the result
        plt.imshow(img, cmap='gray')
        plt.axis('off')  # Hide the axes for better display
        plt.show()
    else:
        print("No circles were detected.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ensure that circles were detected in the previous cell
if 'circles' in locals() and circles is not None and len(circles) >= 2:
    # Get the first two circles
    circle1 = circles[0]
    circle2 = circles[1]
    
    # Get coordinates (x, y) and radius (r) of the two circles
    x1, y1, r1 = circle1
    x2, y2, r2 = circle2

    # Ensure that the circles are positioned in the right order (left to right)
    if x1 > x2:
        x1, y1, r1, x2, y2, r2 = x2, y2, r2, x1, y1, r1

    # Define the bounding box between the two circles
    # Create a region between the right edge of the first circle and the left edge of the second circle
    x_start = x1 + r1
    x_end = x2 - r2
    y_start = min(y1, y2) - max(r1, r2)
    y_end = max(y1, y2) + max(r1, r2)

    # Crop the image using the bounding box defined above
    cropped_img = img[y_start:y_end, x_start:x_end]

    # Plot the cropped region between the circles
    plt.imshow(cropped_img, cmap='gray')
    plt.title("Region between two detected circles")
    plt.axis('off')
    plt.show()
else:
    print("Not enough circles were detected for this operation.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Görüntüyü gri tonlama olarak yükle
img_path = 'Test_images/man.jpg'  # Görüntü yolunu buraya girin
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print(f"Error: Image not found at {img_path}")
else:
    # Gürültüyü azaltmak için Gauss bulanıklığı uygulayın
    blurred_img = cv2.GaussianBlur(img, (21, 21), 0)
    
    # Kenar algılaması yapmak için Canny algoritmasını kullanın
    edges = cv2.Canny(blurred_img, 100, 200)
    
    # Kenarları bulduktan sonra konturları çıkarın
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # Her kontur için bir elips fit etmeye çalışın
    for contour in contours:
        # Konturun nokta sayısı yeterli mi kontrol et
        if len(contour) >= 5:  # Elips fitlemek için en az 5 nokta gerekli
            # Elipsi fit et
            ellipse = cv2.fitEllipse(contour)
            
            # Elipsi görüntüye çiz
            cv2.ellipse(img, ellipse, (0, 255, 0), 2)  # Elipsi yeşil renkte çizin
    
    # Sonuçları göster
    plt.imshow(img, cmap='gray')
    plt.axis('off')  # Eksenleri gizleyin
    plt.show()
```
