```python
import numpy
import matplotlib.pyplot as plt
import cv2 
```

```python
# make a figure from squares wite and black one white one black repeating for 4 times, every time the size of the square is 10x10 pixels

# create a white square
white = numpy.ones((10,10), dtype=numpy.uint8)*255
# create a black square
black = numpy.zeros((10,10), dtype=numpy.uint8)

# create a figure
fig = numpy.concatenate((numpy.concatenate((white, black), axis=1), numpy.concatenate((black, white), axis=1)), axis=0) 
fig = numpy.concatenate((fig, fig), axis=1)
fig_squares = numpy.concatenate((fig, fig), axis=0)

# show the figure
plt.imshow(fig_squares, cmap='gray')
plt.show()
```

```python
# plot the histogram of the figure
plt.hist(fig_squares.ravel(), 256, [0,256])
plt.show()
```

```python
# blur the figure with a 3x3 kernel
kernel = numpy.ones((3,3), numpy.float32)/9
fig_squares_blur = cv2.filter2D(fig_squares, -1, kernel)

# show the blurred figure
plt.imshow(fig_squares_blur, cmap='gray')
plt.show()
```

```python
plt.hist(fig_squares_blur.ravel(), 256, [0,256])
plt.show()
```

```python
# plot all the figures in subplots
plt.subplot(2,2,1)
plt.imshow(fig_squares, cmap='gray')
plt.title('Original')
plt.subplot(2,2,2)
plt.hist(fig_squares.ravel(), 256, [0,256])
plt.title('Histogram Original')
plt.subplot(2,2,3)
plt.imshow(fig_squares_blur, cmap='gray')
plt.title('Blurred')
plt.subplot(2,2,4)
plt.hist(fig_squares_blur.ravel(), 256, [0,256])
plt.title('Histogram Blurred')
plt.tight_layout()
plt.show()
```

```python
# make a half white and half black figure with the same size
white = numpy.ones((fig_squares.shape[0], int(fig_squares.shape[1]/2)), dtype=numpy.uint8)*255
black = numpy.zeros((fig_squares.shape[0], int(fig_squares.shape[1]/2)), dtype=numpy.uint8)

fig_half = numpy.concatenate((white, black), axis=1)

# show the figure
plt.imshow(fig_half, cmap='gray')
plt.show()
```

```python
# plot the histogram of the figure
plt.hist(fig_half.ravel(), 256, [0,256])
plt.show()
```

```python
#blur the figure with a 3x3 kernel
fig_half_blur = cv2.filter2D(fig_half, -1, kernel)

# show the blurred figure and the histogram
plt.imshow(fig_half_blur, cmap='gray')
plt.show()

plt.hist(fig_half_blur.ravel(), 256, [0,256])
plt.show()
```

```python
# plot all the figures in subplots
plt.subplot(2,2,1)
plt.imshow(fig_half, cmap='gray')
plt.title('Original')
plt.subplot(2,2,2)
plt.hist(fig_half.ravel(), 256, [0,256])
plt.title('Histogram Original')
plt.subplot(2,2,3)
plt.imshow(fig_half_blur, cmap='gray')
plt.title('Blurred')
plt.subplot(2,2,4)
plt.hist(fig_half_blur.ravel(), 256, [0,256])
plt.title('Histogram Blurred')
plt.tight_layout()
plt.show()
```
